###### Sequentielle Farbkarten

```python
# --- Funktion zur Generierung der Daten (unverändert) ---
def generate_yearly_temp_data(year=2024):
    """
    Generiert einen zufälligen Datensatz mit täglichen Temperaturen für ein ganzes Jahr,
    realistisch angenähert für eine Region in Mitteleuropa.

    Args:
        year (int): Das Jahr, für das die Daten generiert werden sollen.

    Returns:
        pd.DataFrame: Ein Pandas DataFrame mit den Spalten 'Datum' und 'Temperatur_C'.
    """
    data = []

    # Durchschnittliche Monatstemperaturen und Schwankungsbereiche (für Mitteleuropa, z.B. Deutschland)
    monthly_temp_profiles = {
        1: (1, 3),   # Januar
        2: (2, 3),   # Februar
        3: (6, 4),   # März
        4: (10, 5),  # April
        5: (15, 4),  # Mai
        6: (19, 5),  # Juni
        7: (22, 4),  # Juli
        8: (21, 4),  # August
        9: (16, 4),  # September
        10: (11, 4), # Oktober
        11: (6, 3),  # November
        12: (2, 3)   # Dezember
    }

    current_date = date(year, 1, 1)
    num_days_in_year = 366 if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0) else 365

    for _ in range(num_days_in_year):
        month = current_date.month
        base_temp, temp_range = monthly_temp_profiles.get(month, (10, 5))
        temperature = round(base_temp + random.uniform(-temp_range, temp_range), 1)

        data.append({
            "Datum": current_date,
            "Temperatur_C": temperature # Beachten Sie den Spaltennamen 'Temperatur_C'
        })

        current_date += timedelta(days=1)

    df = pd.DataFrame(data)
    df['Datum'] = pd.to_datetime(df['Datum'])
    return df

# --- Beispiel-Aufruf und Ausgabe ---
# Generiere Daten für das Jahr 2024
yearly_temp_df = generate_yearly_temp_data(year=2024)

print("--- Erster und letzter Teil des generierten Datensatzes (Pandas DataFrame) ---")
print(yearly_temp_df.head())
print("\n...")
print(yearly_temp_df.tail())

print(f"\nGesamtzahl der Tage im Datensatz: {len(yearly_temp_df)}")
print("\n--- DataFrame Info ---")
yearly_temp_df.info()

print("\n--- Deskriptive Statistik der Temperaturen ---")
print(yearly_temp_df['Temperatur_C'].describe())


# --- Heatmap-Code, angepasst an 'yearly_temp_df' und 'Temperatur_C' ---

# Step 1: Extract month and day information from the date column
# 'temperature_data' wird zu 'yearly_temp_df'
calendar_data = yearly_temp_df.copy()
# 'Date' ist jetzt 'Datum'
calendar_data['Month'] = calendar_data['Datum'].dt.strftime('%B')
calendar_data['Day'] = calendar_data['Datum'].dt.day

# Step 2: Define the order of months for a natural calendar sequence
month_names = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

# Convert the 'Month' column to a categorical type with the specified order
calendar_data['Month'] = pd.Categorical(calendar_data['Month'], categories=month_names, ordered=True)

# Step 3: Pivot the data to create a structured table for the heatmap
# 'Temperature' wird zu 'Temperatur_C'
monthly_temperature_table = calendar_data.pivot(index='Month', columns='Day', values='Temperatur_C')

# Step 4: Set up the plot for the heatmap
plt.figure(figsize=(12, 8))

# Define a custom normalization for temperature range (optional)
norm = mcolors.Normalize(vmin=-5, vmax=35) # Angepasster vmin/vmax für realistischere Temperaturen in Mitteleuropa

# Create the heatmap with a sequential colormap
heatmap = plt.imshow(monthly_temperature_table, cmap='YlOrRd', aspect='auto', norm=norm)

# Add color bar to represent temperature
colorbar = plt.colorbar(heatmap, orientation='horizontal')
colorbar.set_label('Temperature (°C)', labelpad=10)

# Label the axes
plt.xlabel('Day of the Month')
plt.ylabel('Month')

# Set y-ticks to display month names instead of numbers
plt.yticks(ticks=np.arange(len(month_names)), labels=month_names)

# Add a title to the heatmap
plt.title('Monatliche Temperatur Heatmap (Sequentielle Colormap)')

# Display the plot
plt.grid(False) # Gitterlinien ausblenden für eine saubere Heatmap
plt.show()
```

###### Divergierende Farbkarten

```python
def generate_random_monthly_profit_loss_data():
    """
    Generiert einen zufälligen DataFrame mit monatlichen Gewinn/Verlust-Daten für ein Jahr.
    Die Daten simulieren typische Geschäftsschwankungen mit Gewinn und Verlust.

    Returns:
        pd.DataFrame: Ein DataFrame mit den Spalten 'Month' und 'Profit/Loss ($)'.
                      'Month' ist hier eine reguläre Spalte (nicht der Index).
    """
    # Liste der Monatsnamen in der richtigen Reihenfolge
    month_names = ['January', 'February', 'March', 'April', 'May', 'June',
                   'July', 'August', 'September', 'October', 'November', 'December']

    profit_loss_values = [] # Liste, um die generierten Werte zu speichern

    # Simulation von Gewinn/Verlust für jeden Monat
    for i, month in enumerate(month_names):
        # Zufällige Gewinn/Verlust-Werte generieren.
        # Die Spannen wurden angepasst, um Vielfalt zu schaffen und Verluste zu ermöglichen.
        # Beispiel: Kältere/ruhigere Monate könnten eher Verluste oder geringere Gewinne haben,
        # wärmere/geschäftigere Monate höhere Gewinne.
        if month in ['January', 'February', 'November', 'December']: # Wintermonate
            value = random.randint(-7000, 5000) # Höhere Chance auf Verlust oder geringen Gewinn
        elif month in ['March', 'April', 'September', 'October']: # Übergangsmonate
            value = random.randint(-2000, 10000) # Ausgeglichenere Spanne
        else: # Sommermonate
            value = random.randint(500, 18000) # Meist Gewinne

        profit_loss_values.append(value) # Wert zur Liste hinzufügen

    # Erstelle den Pandas DataFrame
    data_df = pd.DataFrame({
        'Month': month_names,          # Spalte für die Monatsnamen
        'Profit/Loss ($)': profit_loss_values # Spalte für die Gewinn/Verlust-Werte
    })

    # WICHTIG: Die 'Month'-Spalte als kategorischen Typ mit der definierten Reihenfolge setzen.
    # Das stellt sicher, dass die Monate in der Heatmap in der korrekten Reihenfolge erscheinen
    # und nicht alphabetisch sortiert werden.
    data_df['Month'] = pd.Categorical(data_df['Month'], categories=month_names, ordered=True)

    return data_df # Den fertigen DataFrame zurückgeben

# --- Hauptausführung des Skripts ---
if __name__ == "__main__":
    # Generiere den zufälligen Datensatz
    # 'monthly_profit_loss_data' ist hier ein DataFrame mit 'Month' als regulärer Spalte.
    monthly_profit_loss_data = generate_random_monthly_profit_loss_data()

    print("--- Generierter monatlicher Gewinn/Verlust Datensatz (vor Indexsetzung) ---")
    print(monthly_profit_loss_data)

    # --- HIER beginnt der von Ihnen bereitgestellte Heatmap-Code ---

    # Set 'Month' as the index.
    # Dies transformiert den DataFrame, sodass 'Month' zur Zeilenbeschriftung wird.
    monthly_profit_loss_data.set_index('Month', inplace=True)

    print("\n--- Datensatz nach Indexsetzung (zur Heatmap-Vorbereitung) ---")
    print(monthly_profit_loss_data)


    # Set the figure size for the plot
    plt.figure(figsize=(10, 3)) # Die Größe wurde leicht angepasst für eine bessere Darstellung

    # Create a diverging heatmap using Seaborn
    # 'monthly_profit_loss_data.T' transponiert den DataFrame.
    # Das ist wichtig, damit die Monate auf der X-Achse (Spalten) und
    # die 'Profit/Loss ($)' Beschriftung auf der Y-Achse (Index) der Heatmap erscheint.
    sns.heatmap(monthly_profit_loss_data.T,
                cmap='RdYlGn',     # Farbkarte: Rot für Verlust (negativ), Gelb für Null, Grün für Gewinn (positiv)
                center=0,          # Die Farbkarte wird um den Wert 0 zentriert
                annot=True,        # Zeigt die numerischen Werte direkt in den Zellen an
                fmt='.0f',         # Formatiert die angezeigten Werte als ganze Zahlen (0 Nachkommastellen)
                cbar_kws={'label': 'Profit/Loss ($)'}) # Beschriftung für die Farbleiste

    # Add labels for the axes
    plt.xlabel('Month') # Beschriftung der X-Achse
    plt.ylabel('')      # Y-Achsen-Beschriftung bleibt leer, da der transponierte Index
                        # (die Spalte 'Profit/Loss ($)') automatisch als Label dient

    # Add a title to the heatmap
    plt.title('Profit/Loss Heatmap (Diverging Colormap)')

    # Display the plot
    plt.show()
```

###### Zyklische Farbkarten

```python
# Create data for a sine wave
x = np.linspace(0, 2 * np.pi, 100)  # X values from 0 to 2π (one complete cycle)
y = np.sin(x)  # Y values (sine of X)

# Create a figure and axis
plt.figure(figsize=(10, 6))

# Create a scatter plot with a cyclic colormap
# Color points by their position in the sine wave cycle
points = plt.scatter(x, y, c=x, cmap='twilight', edgecolor='black')

# Add a color bar to show the phase of the wave
cbar = plt.colorbar(points)
cbar.set_label('Wave Phase (Radians)', rotation=270, labelpad=15)

# Label the axes
plt.xlabel('Angle (Radians)')
plt.ylabel('Sine Value')

# Add a title
plt.title('Sine Wave with Cyclic Colormap (Twilight)')

# Display the plot
plt.grid(True)
plt.show()
```

###### Wahrnehmungsmäßig einheitliche Farbkarten

```python
# Generate synthetic elevation data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2)) * 100  # Elevation data

# Plot the elevation data using the 'plasma' colormap
plt.contourf(X, Y, Z, cmap='plasma')

# Add a color bar
plt.colorbar(label='Elevation (m)')

# Add title and labels
plt.title('Topographic Map (Plasma Colormap)')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')

# Display the plot
plt.show()
```
