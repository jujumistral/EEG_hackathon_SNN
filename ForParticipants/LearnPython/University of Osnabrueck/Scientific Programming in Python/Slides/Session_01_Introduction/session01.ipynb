## Advanced Python
Effectively use Python specific features.

```python
class Fibonacci:
    def __init__(self, end):
        self.v1 = 1
        self.v2 = 0
        self.end = end
        
    def __next__(self):
        self.end -= 1
        if self.end <= 0:
            raise StopIteration
        v = self.v2 + self.v1
        self.v2 = self.v1
        self.v1 = v
        return self.v2
    
    def __iter__(self):
        return self

    
for i in Fibonacci(10):
    print(i, end=' ')
```

## Statistical Modeling with statsmodels
Making statistics easy.

```python
import statsmodels.formula.api as smf

n_samples = 100
intercept = 5
slope = 2

x = np.linspace(0, 10, n_samples)
noise = np.random.normal(size=n_samples)
y = intercept + x * slope + noise
data = pd.DataFrame({'x': x, 'y': y})
fit = smf.ols('y ~ x', data).fit()
fit.summary()
```

## Interactive Data Analysis
Explore your data interactively.

```python
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

def f(slope, intercept):
    fig, ax = plt.subplots()
    x = np.linspace(-10, 10, num=1000)
    ax.plot(x, slope * x + intercept)
    ax.set_ylim(-5, 5)
    plt.show()

interactive_plot = interactive(f, slope=(-2.0, 2.0), intercept=(-3, 3, 0.5))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```

## Classes and objects

A class bundles multiple functions ("methods") that operate on the same data (the "object", that is an "instance" of that class):
* classes are defined using the keyword `class`
* methods take a reference to the object as first argument (usually called `self`)

```python
class Square:
    """Simple class representing a square.
    """
    side = 1
    
    def area(self):
        """Area of the square."""
        return self.side ** 2
    
    def perimeter(self):
        """Perimeter of the square."""
        return 4 * self.side
    
    def grow(self, delta):
        """Grow the square."""
        self.side += delta


square = Square()  # instantiate the class
print("area:", square.side, "; area:", square.area(), "; perimeter:", square.perimeter())

square.grow(2)  # change the square
print("area:", square.side, "; area:", square.area(), "; perimeter:", square.perimeter())
```

### List comprehension

Define lists similar to sets in math:
$$L = \{x^2 \mid x \in N\}$$

```python
R = list(range(0,10))
print(R)

L = [x**2 for x in R]

print(L)
```

List comprehensions can be combined:

```python
L = list(range(5))
product = [(a,b) for a in L for b in L]
print(product)
```

Some loops may be done more compactly by list comprehension:

```python
L2 = []
for pair in product:
    L2.append(sum(pair))
print(L2)
```

```python
# shorter with list comprehension:
L2 = [sum(value) for value in product]
print(L2)
```

## File I/O

Files can be opened using `open`. Provide a mode indicating if you want to read (`"r"`) or write (`"w"`).

Write text to a file (overwrites if it exists)

```python
with open("example.txt", "w") as f:
    f.write("First line\nSecond line\n")
```

Append more text

```python
with open("example.txt", "a") as f:
    f.write("Appended line\n")
```

Read the whole file as one string

```python
with open("example.txt", "r") as f:
    content = f.read()
print("Whole file:\n" + content)
```

Read line-by-line

```python
with open("example.txt", "r") as f:
    for i, line in enumerate(f, start=1):
        print(f"Line {i}: {line.rstrip()}")
```

## Coding style
