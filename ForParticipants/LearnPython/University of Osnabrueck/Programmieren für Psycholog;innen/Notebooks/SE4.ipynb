```python
cars = {
    "Toyota": ["Corolla", "Camry", "Prius", "RAV4"],
    "Volkswagen": ["Golf", "Passat", "Tiguan", "ID.4"],
    "Mercedes-Benz": ["C-Klasse", "E-Klasse", "S-Klasse", "EQS"],
    "BMW": ["3er", "5er", "X5", "i4"],
    "Audi": ["A4", "A6", "Q5", "e-tron GT"],
    "Ford": ["Mustang", "F-150", "Focus", "Explorer"],
    "Honda": ["Civic", "Accord", "CR-V", "HR-V"],
    "Hyundai": ["Elantra", "Sonata", "Tucson", "Kona"],
    "Nissan": ["Altima", "Sentra", "Maxima", "Titan"],
    "Opel": ["Corsa", "Astra", "Mokka", "Grandland"],
    "Jeep": ["Wrangler", "Cherokee", "Grand Cherokee", "Compass"],
    "Lexus": ["ES", "RX", "NX", "UX"],
    "Porsche": ["911", "718", "Cayenne", "Macan"],
    "Tesla": ["Model 3", "Model Y", "Model S", "Model X"],
    "Mazda": ["Mazda3", "Mazda6", "CX-5", "CX-9"],
    "Volvo": ["S60", "V60", "XC60", "XC90"],
    "Mini": ["Cooper", "Countryman", "Clubman"],
    "Land Rover": ["Range Rover", "Discovery", "Defender", "Evoque"],
    "Jaguar": ["XE", "XF", "F-PACE", "I-PACE"],
    "Aston Martin": ["Vantage", "DB11", "DBS", "DBX"],
    "Maserati": ["Quattroporte", "Ghibli", "Levante", "MC20"],
    "Lamborghini": ["Aventador", "Huracán", "Urus"],
    "Ferrari": ["812 Superfast", "F8 Tributo", "SF90 Stradale", "Roma"],
    "Rolls-Royce": ["Phantom", "Ghost", "Cullinan", "Wraith"],
    "Bentley": ["Continental GT", "Flying Spur", "Bentayga", "Mulsanne"],
    "Bugatti": ["Chiron", "Veyron", "Divo"]
}

def get_car_models_interactive():
    while True: 
        marke = input("Carl: Bitte geben Sie den Namen eines Autoherstellers ein (oder 'exit' zum Beenden): ").strip() # Leerzeichen entfernen

        if marke.lower() == 'exit': # Eingabe normieren
            print(RED + "M: Das Programm wurde auf Ihren Wunsch hin beendet. Auf Wiedersehen!")
            break 

        Marke = marke.title() # Groß- und Kleinschreibung egalisieren

        try:
            modelle = list(cars[Marke]) # Arbeitskopie der Liste nach Keys des Dictonaries
            
            sug = None # Falls keine Vorschläge vorhanden sind ...
            
            # Interaktiver Modellvorschlag 
            while modelle: 
                sug = random.choice(modelle) 
                
                response = input(f"Carl: Handelt es sich um das Modell '{sug}' von {Marke}? (Ja/Nein): ").strip().lower()

                if response == 'ja':
                    print(GREEN + f"Carl: Ausgezeichnet! Das Modell '{sug}' also - eine ausgezeichnte Wahl!")
                    print("-" * 30)
                    break 
                elif response == 'nein':
                    print(f"Carl: Okay, notiere '{sug}' als abgelehnt.")
                    modelle.remove(sug) 
                    if not modelle: # Sofern aus der Liste alle Optionen entfernt wurden ...
                        print(f"Carl: Bitte entschuldigen Sie! Es wurden leider alle Modelle von {Marke} von Ihnen abgelehnt. Wir haben leider keine weiteren Modelle in unseren Datenbanken vermerkt.")
                        print("M: Bitte entschuldigen Sie diesen Fehler unsererseits - versuchen Sie es noch einmal mit einer anderen Marke.")
                        print("-" * 30)
                        break # Innere Schleife verlassen, um neue Herstellerabfrage zu starten (äußere Schleife läuft weiter)
                else:
                    print(RED + "M: Ungültige Eingabe. Bitte geben Sie 'Ja' oder 'Nein' ein.")

        except KeyError:
            print(RED + f"M: Fehler: Der Hersteller '{Marke}' wurde nicht in unserer Datenbank gefunden.")
            print(RED + "M: Bitte überprüfen Sie die Schreibweise oder versuchen Sie einen anderen Hersteller.")
            print("M: Hier noch einmal ein paar der bei uns gelisteten Hersteller (Auswahl):", ", ".join(list(cars.keys())[:5]), "...") # Beispiele für Keys (Marken), gesliced und zu einer Liste zusammengeführt
            print("-" * 30)

        except Exception as e:
            print(RED + f"M: Ein unerwarteter Fehler ist aufgetreten: {e}")
            print("M: Bitte versuchen Sie es erneut oder wenden Sie sich an den Support.")
            print("-" * 30)

if __name__ == "__main__":
    print("M: Willkommen bei WeGuessYourCar! Ihr heutiger Betreuer wird unser Praktikant Carl sein. Wir wünschen Ihnen viel Spaß!")
    print("Carl: Hallo!")
    print("Carl: Dann fangen wir mal an!")
    print()
    
    get_car_models_interactive()
```

• *Schreibe eine **Funktion**, die einen Input-Vektor von beliebiger Länge in
umgekehrter Reihenfolge als Output generiert*

```python
def get_list_of_numbers():
    while True:
        numbers = input("M: Bitte geben Sie eine Liste von (ganzen) Zahlen zwischen 0 und 9999 ein (durch Komma getrennt): ")
        
        # String an den Kommas teilen und entfernen der Leerzeichen
        num_strings = [s.strip() for s in numbers.split(',')] # Leerzeichen entfernen und an Kommata trennen
        
        parsed_numbers = []
        all_valid = True 

        for num_str in num_strings:
            try:
                num = int(num_str)
                if 0 <= num <= 9999:
                    parsed_numbers.append(num)
                else:
                    print(RED + f"M: Entschuldigen Sie, aber '{num_str}' ist keine gültige Eingabe. Bitte geben Sie nur Zahlen zwischen 0 und 9999 ein." + RESET)
                    all_valid = False
                    break 
            except ValueError:
                print(RED + f"M: Entschuldigen Sie, aber '{num_str}' ist keine gültige Eingabe. Bitte geben Sie ausschließlich ganze Zahlen ein." + RESET)
                all_valid = False
                break 
        
        if all_valid:
            return parsed_numbers 

liste = []

if __name__ == "__main__":
    print("M: Willkommen bei UnoReverse! Wir wünschen Ihnen viel Spaß!")
    print()
    
    liste = get_list_of_numbers() 

    liste.reverse()
    print(GREEEN + f"M: Ihre Liste lautet also {liste} - sehr interessant!")
```

• *schreibe eine **Funktion**, die die Fibonacci-Folge der Länge n generiert*

```python
# 1. Funktion zur Berechnung der Fibonacci-Folge
def calculate_fibonacci_sequence(n):
    if n < 0:
        return None 
    elif n == 0:
        return [0]
    elif n == 1:
        return [0, 1]
    else:
        fib_sequence = [0, 1]
        a, b = 0, 1
        for _ in range(2, n + 1): # durchführen der Itterarion n-1 Mal 
            next_fib = a + b
            fib_sequence.append(next_fib)
            a = b
            b = next_fib
        return fib_sequence

# 2. Eingabefunktion
def get_fibonacci_n():
    while True:
        try:
            given_str = input("M: Bitte geben Sie eine nicht-negative ganze Zahl (n) ein, bis zu der wir Ihnen die Fibonacci-Folge ausgeben sollen: ")
            given_n = int(given_str)

            if given_n >= 0: 
                return given_n 
            else:
                print(RED + "M: Ungültige Eingabe. Bitte geben Sie eine nicht-negative ganze Zahl ein." + RESET)
        except ValueError:
            print(RED + "M: Ungültige Eingabe. Bitte geben Sie ausschließlich ganze Zahlen ein." + RESET)

if __name__ == "__main__":
    print("M: Willkommen bei Fibonacci-Numbers!")

    n_value = get_fibonacci_n()

    fib_result_sequence = calculate_fibonacci_sequence(n_value)

    if fib_result_sequence is not None:
        print(GREEN + f"M: Hier folgt Ihre Fibonacci-Folge bis zur Stelle {n_value}: {fib_result_sequence}")
    else:
        print(RED + "M: Es konnte keine Fibonacci-Folge für die angegebene Stelle berechnet werden." + RESET)
```
